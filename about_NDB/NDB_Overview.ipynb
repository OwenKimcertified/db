# Network DataBase (MySQL)

NDB 는 nothing-shared 공유하지 않는 in - memory clustering 이다.

disk I/O 보다 빠르며, 독립적이기 때문에 vertical scale - out 이 쉽고, 

high availability 를 보장하여 단일 지점 장애 (SPOF) 를 방지할 수 있다..

SPOF (Single Point Of Failure) 를 방지하기 위해 각각의 노드가 독립적인

디스크와 메모리를 가져야하고, 그로 인해 nothing - shared 가 필요한 것이다.

### Subnet mask

--subnet=0.0.0.0/16 에서 ~/16 은 서브넷 마스크.

`서브넷 마스크는 IP 주소의 범위를 설정하는데 사용하고`

`IP 주소를 네트워크 / 호스트 부분으로 구분짓는다.`

16은 네트워크 비트 수를 의미하며, 첫 번째 16비트가 네트워크 주소를 나타내고 나머지 비트가 호스트 주소임.

예를 들어 192.168.0.0/16 은 192.168.0.0 에서 192.168.255.255 까지의 범위를 가질 수 있고,

65,000 개 정도의 주소를 가질 수 있음. 따라서 `192.168 까지는 네트워크의 주소`, `그 이후의 자리는 호스트 주소를 나타낸다`.
