### Docker lifecycle

도커 컨테이너에 의해 생성된 데이터들은 컨테이너가 삭제되면 같이 삭제됨.

도커의 컨테이너는 격리된 공간(isolated)이기 때문에, 미리 조치하지 않으면 

데이터가 사라지는 경험을 할 수도 있다.

![test](./imgs/7.png)

위의 그림과 같이 바인드 마운트와 볼륨에는 범위의 차이가 있다.

1. bind mount

바인드 마운트는 호스트 시스템(로컬 or etc)의 특정 경로를(로컬 파일 시스템) 도커의 특정 경로(도커 파일 시스템)에 연결함.

그로 인해 호스트의 파일 시스템에 있는 파일 및 디렉토리를 컨테이너에 마운트할 수 있게된다.

호스트 파일 시스템이 변경되면 도커의 컨테이너 파일 시스템에 반영되기 때문에 유연하지만 보안에 취약점이 있다.

2. `volume`

볼륨은 호스트 파일 시스템의 일부를 도커가 관리하는 데이터의 저장 영역.

여러 컨테이너가 한 볼륨을 참조할 수 있고 컨테이너를 종료해도 볼륨 자체는 도커에서 계속 활용할 수 있다.

공식 문서에서는 volume 활용을 권장하며 1번의 바인드 마운트보다 성능이 좋다라고 말한다.

api 로 관리할 수 있으며 볼륨의 내용을 암호화 할 수 있다.
